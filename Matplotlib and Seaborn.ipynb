
# Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt

xarr=np.array(['Python','C','C++','JAVA'])
yarr=np.array([90,70,60,85])
plt.xlabel('Languages', fontsize=15)
plt.ylabel('Popularity', fontsize=15)
col=np.array(['g','c','m','b'])
plt.bar(xarr, yarr, width=0.6, align='center', color=col, edgecolor='r', linewidth=2, alpha=0.8)
plt.title('Popularity of Programming Languages', fontsize=20)
plt.show()
```

```python
# Multiple Bar Graphs

xarr=np.array(['Germany','Britain','India','USA'])
plt.xlabel('Countries')
yarr=np.array([90,50,60,70])
zarr=np.array([50,40,20,30])

width=0.4
p=np.arange(len(xarr))  # array[0,1,2,3]
p1=[j+width for j in p]
plt.ylabel('Imported oil supplies(in billion tons)')
plt.bar(p, yarr, width, color='c', label='2024')
plt.bar(p1, zarr, width, color='g', label='2010')
plt.xticks(p+width/2, xarr, rotation=50)
plt.legend()
# to show both the graphs together
# we will create a list with parameters (index number)
# since both have same width we can keep the width paarameter  outside the function
plt.show()
```

```python
xarr=np.array(['Virat','Rohit','Cummins','Root'])
yarr=np.array([136,120,50,80])
zarr=np.array([98,100,40,60])
plt.xlabel('Cricketers', fontsize=12)
plt.ylabel('Stats', fontsize=12)
width=0.3
p=np.arange(len(xarr))   # array([0,1,2,3])
p1=[j+width for j in p]


plt.bar(p, yarr, width, align='center')
plt.bar(p1, zarr, width, align='center')
plt.xticks(p+width/2, xarr, rotation=40)

# plt.xtixks(position, x axis, rotation)
plt.legend(['Runs scored', 'Strike Rate'])
plt.title('Statistics of Players', fontsize=15)
plt.show()
```

# Pie Chart/Plot

```python
# using plt.pie()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['C','C++','Java','Python']   # labels for pie plots
ex=[0.3,0.0,0.0,0.0]
col=['r','b','g','y']

plt.pie(x, labels=y, explode=ex, autopct='%0.1f%%', shadow=True, radius=1.1, labeldistance=1.2, startangle=0,
       textprops={'fontsize':15}, counterclock=False, wedgeprops={'linewidth':4},
       rotatelabels=False)
plt.title('Pie Plot')
plt.legend(loc='upper left')
plt.show()



# using labels parameter we can give name to each part of pie plot
# to show a particular part differently, we use explode parameter
# we can change the colors of pies using colors parameter
# to get the percentage of ecah plot we use autopct (%0.1f%%)parameter
# we can use shadow parameter to give shadow
# we can increase the radius of pie using radius parameter
# we can change the label distance  using labeldistance property
# we can change the starting point of pie plot --> using startingangle parameter
# to change the font size of labels , we do this using textprops parameter(we give values in the form of dictionary)
# we can change the font color using fontcolor property
# if we wnat to change the direction of labels ,we use counterclock property (by default it is anticlockwise)
# we can change the line width using wedgeprops parameter --> linewidth property
# we can change the center using center property
# we cam rotate labels using rotatelabels parameter
# use legend
```

# Seaborn Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var1=[1,2,3,4,5,6,7]
var2=[2,3,4,1,5,6,7]

plt.plot(var1,var2)
plt.show()
```

```python
var1=[1,2,3,4,5,6,7]
var2=[2,3,4,1,5,6,7]
# create a dataframe by making a dictionary out of two
df=pd.DataFrame({'var_1':var1,
                'var_2':var2})

sns.lineplot(x='var_1', y='var_2', data=df)
plt.show()

# Here if we plot the line plot using dataframe
# then we can directly show the x and y axis labels
```

```python
var1=[1,2,3,4,5,6,7]
var2=[2,3,4,1,5,6,7]

# Without creating a dataframe out of lists
sns.lineplot(x=var1, y=var2)
plt.show()
```

```python
# For large amount of data consider using a dataset
dset=sns.load_dataset("penguins").head(20)
dset
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=dset, hue='sex', style='sex', palette='viridis', markers=['o','>'], dashes=False, legend=False)
plt.grid()
plt.title('Seaborn Line Plot')
plt.show()
```

```python
# if we want to create a graph wrt the most popular series in dataframe
# then we use 'hue' parameter

# we can customise the size using 'size' parameter
# to make changes in the types of lines in legend we make us of 'style' parameter
# Insted of using cmap property, we make use of 'palette' parameter
# To show markers at the corner points, we make use of 'markers' parameter
# If we want to rmove the dashed lines , then we make use of 'dashes'=False parameter
# To remove legend, we make the 'legend' parameter False
```

# Bar Plot

```python
# using sns.barplot()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dset=sns.load_dataset("penguins").head(50)
dset
```

```python
ord=['Dream','Torgersen','Biscoe']
sns.set(style='darkgrid')
sns.barplot(x='island' , y='bill_length_mm', data=dset, hue='sex', order=ord, ci=100,
            orient='v', saturation=100, errcolor='g', errwidth=1,
           capsize=0.2, dodge=True, alpha=1)

plt.show()

# Using 'hue' parameter, we can easily create multiple bar plot
# we can change the order of data on x axis using 'order' parameter
# we can change the order of legend using 'hue_order' parameter
# to change the interval, we use 'ci' parameter (0-100) range
# we can change the orienattion using 'orient' parameter
# for change colour, we use 'color' parameter
# we can also use 'palette' parameter
# Here instead of alpha parameter we use 'saturation' parameter
# we can change the color of error lines using 'errcolor' property
# and error width using 'errwidth' prop
# to show end point we can use 'capsize' property
# we can superimpose the plots using 'dodge' parameter
# using seaborn, we can create a grid as :
# sns.set(style='darkgrid')
```

```python
sns.barplot(x='island' , y='bill_length_mm', data=dset, hue='sex', order=ord, hue_order=['FEMALE','MALE'])
plt.show()
```

```python
sns.barplot(x='bill_depth_mm', y='bill_length_mm', data=dset, orient='h')
plt.show()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=dset, color='b')
plt.show()
```

```python
# Heatmap

x=np.linspace(1,40,(20)).reshape(4,5)
print(x)
```

```python
val={
    'fontsize':20,
    'color':'w'
}
data=sns.heatmap(x, vmin=0, vmax=40, annot=True,
            cmap='viridis', annot_kws=val,
           linewidth=4, linecolor='k', cbar=False,
           xticklabels=False,yticklabels=False)
data.set(xlabel='Seaborn', ylabel='Python')
plt.show()


# To separate the boxes we have linewidth property
# boundary color can be cahnged using linecolor
# to remove color bar, we do cbar=False
# we can remove xticks and yticks by xticklabels=False
```

```python
y=np.linspace(1,10,(10)).reshape(2,5)
y
```

```python
# To change the values in boxes in strings
# we will change annot parameter and format as string
arr=np.array([
    ['c0','c1','c2','c3','c4'],
    ['d0','d1','d2','d3','d4']
])


sns.heatmap(y, vmin=0, vmax=10,annot=arr, fmt='s')
plt.show()
```
